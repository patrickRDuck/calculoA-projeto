# üìò Quest√£o 1 ‚Äî Limites

Em um laborat√≥rio de inform√°tica, pesquisadores est√£o analisando o comportamento t√©rmico de um processador sob carga cont√≠nua. Ap√≥s um certo tempo de uso intenso, a temperatura do processador tende a se estabilizar, aproximando-se de um valor limite devido √†s a√ß√µes do sistema de resfriamento.

Para representar esse fen√¥meno, considere uma fun√ß√£o que descreve a temperatura T(t) da CPU ao longo do tempo t, durante uma simula√ß√£o de estresse t√©rmico.

Com base nesse cen√°rio, as quest√µes abaixo acompanhar√£o o comportamento da fun√ß√£o √† medida que t se aproxima de valores-chave.

```python
# Quest√£o 1A
limite1A = smp.Limit((x**2 - 9)/ (x-3), x, 3)

print("Express√£o do Limite 1A com x tendendo a 3:")
display(limite1A)

print("Resultado do Limite:")
limite1A.doit() 
```

```python
# Quest√£o 1B
limite1B = smp.Limit((smp.sqrt(x) - 2)/ (x-4), x, 4)

print("Express√£o do Limite 1B com x tendendo a 4:")
display(limite1B)

print("Resultado do Limite:")
limite1B.doit() 
```

```python
# Quest√£o 1C
limite1C = smp.Limit((1 - smp.cos(x))/ (x**2), x, 0)

print("Express√£o do Limite 1C com x tendendo a 0:")
display(limite1C)

print("Resultado do Limite:")
limite1C.doit() 
```

```python
# Quest√£o 1D
limite1D = smp.Limit((E**x - 1) / x, x, 0)

print("Express√£o do Limite 1D com x tendendo a 0:")
display(limite1D)

print("Resultado do Limite:")
limite1D.doit() 
```
