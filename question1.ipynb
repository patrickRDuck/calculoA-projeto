# üìò Quest√£o 1 ‚Äî Limites

Em um laborat√≥rio de inform√°tica, pesquisadores est√£o analisando o comportamento t√©rmico de um processador sob carga cont√≠nua. Ap√≥s um certo tempo de uso intenso, a temperatura do processador tende a se estabilizar, aproximando-se de um valor limite devido √†s a√ß√µes do sistema de resfriamento.

Para representar esse fen√¥meno, considere uma fun√ß√£o que descreve a temperatura T(t) da CPU ao longo do tempo t, durante uma simula√ß√£o de estresse t√©rmico.

Com base nesse cen√°rio, as quest√µes abaixo acompanhar√£o o comportamento da fun√ß√£o √† medida que t se aproxima de valores-chave.

```python
# Quest√£o 1A
limite1A = smp.Limit((x**2 - 9)/ (x-3), x, 3)

print("Express√£o do Limite 1A com x tendendo a 3:")
display(limite1A)

print("Resultado do Limite:")
limite1A.doit() 
```

```python
# Quest√£o 1B
limite1B = smp.Limit((smp.sqrt(x) - 2)/ (x-4), x, 4)

print("Express√£o do Limite 1B com x tendendo a 4:")
display(limite1B)

print("Resultado do Limite:")   
limite1B.doit() 
```

```python
# Quest√£o 1C
limite1C = smp.Limit((1 - smp.cos(x))/ (x**2), x, 0)

print("Express√£o do Limite 1C com x tendendo a 0:")
display(limite1C)

print("Resultado do Limite:")
limite1C.doit() 
```

```python
# Quest√£o 1D
limite1D = smp.Limit((E**x - 1) / x, x, 0)

print("Express√£o do Limite 1D com x tendendo a 0:")
display(limite1D)

print("Resultado do Limite:")
limite1D.doit() 
```

```python
# Quest√£o 1E
limite1E = smp.Limit((3 * x**2 - 5*x + 1) / (x ** 2 + 2), x, smp.oo, dir="+")

print("Express√£o do Limite 1E com x tendendo ao infinito:")
display(limite1E)

print("Resultado do Limite:")
limite1E.doit() 
```

```python
# Quest√£o 1F
limite1F = smp.Limit((2 * x**3 + 7*x**2 - x + 4) / (-x ** 3 + 5), x, smp.oo, dir="-")

print("Express√£o do Limite 1F com x tendendo ao -infinito:")
display(limite1F)

print("Resultado do Limite:")
limite1F.doit() 
```

```python
# Quest√£o 1G
limite1G = smp.Limit((x-3) / smp.Abs(x-3), x, 3, dir="-")

print("Express√£o do Limite 1G com x tendendo a 3-:")
display(limite1G)

print("Resultado do Limite:")
limite1G.doit() 
```

```python
# Quest√£o 1H
limite1H = smp.Limit((x-2) / smp.Abs(x-2), x, 2, dir="+")

print("Express√£o do Limite 1H com x tendendo a 2+:")
display(limite1H)

print("Resultado do Limite:")
limite1H.doit() 
```

```python
# Quest√£o 1I
limite1I = smp.Limit((x**2 - 16) / (x-4), x, 4)

print("Express√£o do Limite 1I com x tendendo a 4+:")
display(limite1I)

print("Resultado do Limite:")
limite1I.doit() 
```

```python
# Quest√£o 1J
limite1J = smp.Limit((5 * x**3 - 2*x + 1) / (x**3 + 4*x), x, smp.oo)

print("Express√£o do Limite 1J com x tendendo ao infinito+:")
display(limite1J)

print("Resultado do Limite:")
limite1J.doit() 
```

```python
# Quest√£o 1K
limite1K = smp.Limit(x * E**-x, x, smp.oo)

print("Express√£o do Limite 1K com x tendendo ao infinito+:")
display(limite1K)

print("Resultado do Limite:")
limite1K.doit() 
```

```python
# Quest√£o 1L
limite1L = smp.Limit(smp.sqrt(x**2 + x) - x, x, smp.oo)

print("Express√£o do Limite 1L com x tendendo ao infinito+:")
display(limite1L)

print("Resultado do Limite:")
limite1L.doit() 
```

```python
# Quest√£o 1M
limite1M = smp.Limit((1 + 2/x)**x , x, smp.oo)

print("Express√£o do Limite 1M com x tendendo ao infinito+:")
display(limite1M)

print("Resultado do Limite:")
limite1M.doit() 
```

```python
# Quest√£o 1N
limite1N = smp.Limit(x**x , x, 0)

print("Express√£o do Limite 1N com x tendendo a 0+:")
display(limite1N)

print("Resultado do Limite:")
limite1N.doit() 
```
