```python
import sympy as smp
from sympy import *

from IPython.display import display

x, y = smp.symbols('x y')
```

## Calculadora de limites


```python
#escreva aqui a expressão matemática que você deseja calcular o limite
expressao_de_entrada_limite = (x**2 - 9)/ (x-3)

# Escreva aqui para onde o limite tende
tendencia_do_limite = 3

# Escreva aqui a direção do limite, use "+" ou "-"
direcao_do_limite = "+"

limite = smp.Limit(expressao_de_entrada_limite, x, tendencia_do_limite, dir=direcao_do_limite)

print("Expressão do Limite:")
display(limite)

print("Resultado do Limite:")
limite.doit() 
```

## Calculadora de derivada

```python
#escreva aqui a expressão matemática que você deseja calcular a derivada
expressao_de_entrada_derivada = (2*x**4-4*x**2)


derivada = smp.Derivative(expressao_de_entrada_derivada, x)

print('Expressão da derivada:')
display(derivada)

print(f'Resultado da derivação é igual a:')
display(derivada.doit())
```

# Calculadora de Integral 

```python
# Escreva aqui a expressão matemática que você deseja calcular a integral
expressao_de_entrada_integral = 3*x**2 + 5*x + smp.sin(x)

integral = smp.Integral(expressao_de_entrada_integral, x)

print("Expressão da integral:")
display(integral)

print("Resultado da integral:")
integral.doit() 
```
