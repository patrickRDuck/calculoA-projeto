# üìò Quest√£o 3 ‚Äî Integrais

Uma empresa de infraestrutura de TI monitora o tr√°fego de dados de um servidor que atende requisi√ß√µes de milhares de usu√°rios simult√¢neos. A taxa de transfer√™ncia de dados √© dada por uma fun√ß√£o R(t), que representa quantos megabytes por segundo s√£o transmitidos no instante t.

Para determinar o volume total de dados enviados durante um intervalo espec√≠fico ‚Äî como uma janela de pico de acessos ou um per√≠odo de baixa ‚Äî √© necess√°rio calcular a integral da fun√ß√£o de taxa.

Nos itens subsequentes, ser√° determinado o total de dados transferidos em diferentes intervalos, interpretando a integral como o ac√∫mulo do tr√°fego de rede ao longo do tempo.

```python
# Quest√£o 3A
integral3A = smp.Integral(3*x**2 + 5*x + smp.sin(x), x)

print("Express√£o da integral 3A:")
display(integral3A)

print("Resultado da integral:")
integral3A.doit() 
```

```python
# Quest√£o 3B
integral3B = smp.Integral(3*x**2 * smp.sqrt(x**3 + 5), x)

print("Express√£o da integral 3B:")
display(integral3B)

print("Resultado da integral:")
integral3B.doit() 
```

```python
# Quest√£o 3C
integral3C = smp.Integral(x * smp.exp(x), x)

print("Express√£o da integral 3C:")
display(integral3C)

print("Resultado da integral:")
integral3C.doit() 
```
