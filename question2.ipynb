Uma equipe de desenvolvedores est√° monitorando o consumo de mem√≥ria RAM de uma aplica√ß√£o complexa que realiza processamento de dados em tempo real. O uso de mem√≥ria ao longo do tempo √© representado por uma fun√ß√£o M(t), que registra quantos gigabytes est√£o sendo utilizados em cada instante t.

A derivada dessa fun√ß√£o indica a taxa instant√¢nea de crescimento ou libera√ß√£o de mem√≥ria, informa√ß√£o essencial para detectar vazamentos de mem√≥ria, gargalos e picos inesperados durante a execu√ß√£o.

Com base nesse cen√°rio, as quest√µes abaixo mostrar√£o os momentos quando o consumo aumenta rapidamente, se estabiliza ou diminui, utilizando o conceito de taxa de varia√ß√£o instant√¢nea.

```python
import sympy as smp
from sympy import *
```

```python
# respons√°vel por carregar a fun√ß√£o display() do m√≥dulo IPython.display para exibir objetos formatados no Jupyter Notebook.
from IPython.display import display
```

```python
# Inicializa a impress√£o simb√≥lica com suporte a LaTeX (configura√ß√£o para Jupyter Notebook)
smp.init_printing(use_latex='mathjax')
```

```python
# Define symbols
x = smp.symbols('x')
```

```python
#Quest√£o 2A
f = (2*x**4-4*x**2)
derivada2 = smp.diff(f, x)

print('Deriva√ß√£o direta pela f√≥rmula da tabela, n√∫mero 1 (regra da pot√™ncia):')
display(f)
print(f'Resultado da deriva√ß√£o √© igual a:')
display(derivada2)
```

```python
#Quest√£o 2B
f = (3*x**2-1)*(4*x+3)
derivada2 = smp.diff(f, x)

print('Deriva√ß√£o direta pela f√≥rmula da tabela, n√∫mero 2 (regra do produto):')
display(f)
print(f'Resultado da deriva√ß√£o √© igual a:')
display(derivada2)
```

```python
#Quest√£o 2C
f = (-x**3+1)/(3*x-5)
derivada2 = smp.diff(f, x)

print('Deriva√ß√£o direta pela f√≥rmula da tabela, n√∫mero 3 (regra do quociente):')
display(f)
print(f'Resultado da deriva√ß√£o √© igual a:')
display(derivada2)
```

```python
#Quest√£o 2D
f = -3**x
derivada2 = smp.diff(f, x)

print('Deriva√ß√£o direta pela f√≥rmula da tabela, n√∫mero 4:')
display(f)
print(f'Resultado da deriva√ß√£o √© igual a:')
display(derivada2)
```
